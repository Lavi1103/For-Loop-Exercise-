```python
# Question 7
num = int(input("Enter a number: "))

if num <= 1:
    print("Not a prime number")
else:
    for i in range(2, num):
        if num % i == 0:
            print("Not a prime number")
            break
    else:
        print("Prime number")
```

```python
# question 8

numbers = [3, 7, 2, 9, 5]

largest = numbers[0]   
for num in numbers:
    if num > largest:
        largest = num

print("Largest element is:", largest)
```
